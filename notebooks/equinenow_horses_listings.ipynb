##**Diccionario de Datos:**

- **horse_id:** Identificacion del caballo
- **Breed:** Raza
- **Name:** Nombre
- **Gender:** Genero.  en caballos definen su etapa de vida, género y estado reproductivo. Los principales son: Filly (potranca < 4 años), Colt (potro macho < 4 años), Mare (yegua), Stallion (semental/macho entero), Gelding (caballo castrado), Foal (potro recién nacido), Weanling (destetado), Yearling (añojo), Broodmare (yegua de cría), Unborn (no nacido) y Unknown (desconocido).
- **Foal Date:** Fecha de nacimiento del potro/potranca.
- **In Foal:** Si la Yegua está embarazada o preñada (en gestación)
- **Height (hh):**"manos de altura" (hands high en inglés). Es la unidad de medida tradicional utilizada para medir la altura de un equino desde el suelo hasta la cruz (la base del cuello). Una mano equivale a 4 pulgadas o 10.16 cm.
- **Weight (lbs):** Peso corporal del animal medido en libras (libras-peso)
- **Color:** Describe los diferentes colores de pelaje (capas) y patrones de manchas en caballos, incluyendo bases sólidas, diluciones, patrones píos (manchas) y ruanos.
- **Registry:** Incripción oficial de un ejemplar
- **Temperament:** Personalidad innata, comportamiento general y forma de reaccionar ante estímulos del entorno
- **Ad Number:** Número de identificación del caballo
- **Ad Created:** Fecha de publicación
- **Last Update:** Ultima actualiación
- **Location:** Ubicación
- **Price:** Precio
- **Horse Profile:** Perfil del caballo (página web)
- **Skills:** Las habilidades, capacidades físicas y técnicas entrenadas (velocidad, resistencia, salto, doma, trote, galope) que determinan su rendimiento
- **Comments:** Comentarios.
- **Shipping:** Envío.
- **Company Name:** Nombre de la empresa.
- **Company Profile:** Perfil de la empresa.
- **State Bred:** Se refiere a un ejemplar que ha nacido y ha sido registrado en un estado específico de Estados Unidos, cumpliendo con los requisitos de residencia y cría de la asociación de criadores de ese lugar.
- **Markings:** Manchas o zonas de pelo blanco que presentan de forma natural, principalmente en la cabeza y las extremidades, sobre una capa base más oscura
- **Registry Number:** Número de Registro.

```python
#crear copia del df 
df_final = df.copy()
```

```python
#borrar columnas que no sirven 
cols_a_borrar = ['Ad Number', 'Registry Number', 'State Bred']
df_final = df_final.drop(columns=cols_a_borrar, errors='ignore')
```

```python
#limpieza de texto
cols_texto = df_final.select_dtypes(include=['object']).columns
for col in cols_texto:
    df_final[col] = df_final[col].astype(str).str.lower().str.strip()
    #reemplazar nan o none por nulos reales
    df_final[col] = df_final[col].replace({'nan': None, 'none':None, 'sin información': None})
```

```python
#limpieza de $ y coma 
df_final['Price'] = df_final['Price'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False)
```

```python
#forazando números, si falla se coloca NaN
cols_numericas = ['Price', 'Height (hh)', 'Weight (lbs)']
for col in cols_numericas:
    df_final[col] = pd.to_numeric(df_final[col], errors='coerce')
```

```python
#fechas 
cols_fechas = ['Foal Date', 'Ad Created', 'Last Update']
for col in cols_fechas:
    df_final[col] = pd.to_datetime(df_final[col], errors='coerce', format='mixed')
```

```python
#calcular edad
#si tenemos fecha de nacimiento calculamos la edad exacta 
now = pd.Timestamp.now()
df_final['Age'] = (now - df_final['Foal Date']).dt.days / 365
df_final['Age'] = df_final['Age'].round(1)
```

```python
#detectar marcas de valor(cara)
palabras_clave = 'star|blaze|strip|bald|white face|snipe'
# Rellenamos nulos temporales en Markings para poder buscar
df_final['Markings'] = df_final['Markings'].fillna('sin información')
df_final['Has_Face_Markings'] = df_final['Markings'].str.contains(palabras_clave, regex=True).astype(int)
```

```python
# 4.1. Rellenar EDAD y PRECIO usando la Mediana por RAZA 
# (Si es un Pony, usa la edad/precio promedio de Ponys)
cols_a_rellenar = ['Price', 'Height (hh)', 'Weight (lbs)', 'Age']

for col in cols_a_rellenar:
    #intenta rellenar con la mediana de su raza
    df_final[col] = df_final[col].fillna(df_final.groupby('Breed')[col].transform('median'))
    #si la raza es unica y no tiene mediana , usa la mediana global 
    df_final[col] = df_final[col].fillna(df_final[col].median())
```

```python
#rellenar fechas faltantes
# si no hay fecha de actualización , se asume la de creación 
df_final['Last Update'] = df_final['Last Update'].fillna(df_final[col].median())
```

```python
#rellenar texto restante
cols_texto_final = df_final.select_dtypes(include=['object', 'string']).columns
df_final[cols_texto_final] = df_final[cols_texto_final].fillna('sin información')
```

```python
#limpieza final 
df_final = df_final.drop(columns=['Foal Date'], errors='ignore')
```

```python
import re

# Función para detectar si un texto tiene "mugre" (algo que no sea letra, número o espacio simple)
def tiene_caracteres_raros(texto):
    if pd.isna(texto): return False
    # Busca cualquier cosa que NO sea letra, número, espacio, punto, coma o guión
    # Si encuentra un emoji o un asterisco, devuelve True
    return bool(re.search(r'[^a-zA-Z0-9\s\.\,\-ñÑáéíóúÁÉÍÓÚ]', str(texto)))

# Revisamos las columnas de texto
cols_texto = df_final.select_dtypes(include=['object', 'string']).columns

print("--- Filas con caracteres raros detectados ---")
for col in cols_texto:
    # Contamos cuántas filas tienen suciedad
    sucias = df_final[col].apply(tiene_caracteres_raros).sum()
    if sucias > 0:
        print(f"Columna '{col}': {sucias} filas sucias")
        # Mostramos un ejemplo para que veas qué encontró
        ejemplo = df_final[df_final[col].apply(tiene_caracteres_raros)][col].iloc[0]
        print(f"   Ejemplo: {ejemplo}")
```

```python
# Definimos el patrón de lo que SÍ queremos mantener
# [^ ... ] significa "Borra todo lo que NO sea esto"
# a-z0-9: Letras y números
# \s: Espacios
# ñÑáéíóú...: Acentos y eñes (por si acaso)
# \.\,\-\': Puntos, comas, guiones y apóstrofes (para nombres como O'Connor)
patron_a_mantener = r'[^a-zA-Z0-9\sñÑáéíóúÁÉÍÓÚ\.\,\-\']'

print("Limpiando asteriscos y emojis... 🧹")

for col in cols_texto:
    # 1. Reemplazamos todo lo que NO sea el patrón con vacío ''
    df_final[col] = df_final[col].astype(str).str.replace(patron_a_mantener, '', regex=True)
    
    # 2. Quitamos espacios dobles que hayan quedado 
    df_final[col] = df_final[col].str.replace(r'\s+', ' ', regex=True).str.strip()

print("¡Limpieza terminada! ✨")

# Verificación
print(df_final[['Name', 'Comments']].head())
```

```python
# --- TRADUCCIÓN RÁPIDA (Español -> Inglés) 🇺🇸 ---

# 1. Seleccionamos las columnas de texto
cols_texto = df_final.select_dtypes(include=['object', 'string']).columns

# 2. Definimos qué cambiar (incluye variantes por si acaso)
traduccion = {
    'sin información': 'unknown',
    'sin informacion': 'unknown',   # Por si falta la tilde
    'no disponible': 'not provided',
    'consultar': 'ask seller',
    'nan': 'unknown',
    'none': 'unknown'
}

print(f"--- Traduciendo {len(cols_texto)} columnas... ---")

for col in cols_texto:
    # Aseguramos que sea texto y minúsculas
    df_final[col] = df_final[col].astype(str).str.lower().str.strip()
    
    # Aplicamos el cambio
    df_final[col] = df_final[col].replace(traduccion)

print("✅ ¡Listo! Todo ahora dice 'unknown'.")
print("Ejemplo:", df_final[cols_texto[0]].iloc[0]) # Verificación rápida
```

```python
# --- RESCATE DE LINKS ORIGINALES 🔗 ---

# 1. Identifica las columnas que son URL (Links)
# Ajusta esta lista con los nombres exactos de tus columnas de links
cols_links = ['Horse Profile', 'Company Profile'] 

print(f"Restaurando links originales en: {cols_links}...")

# 2. Pasamos los datos "crudos" del original al final
# (Esto sobreescribe la versión limpia/rota con la versión original/funcional)
for col in cols_links:
    if col in df.columns and col in df_final.columns:
        df_final[col] = df[col] # ¡Aquí vuelve la magia original!
        
        # 3. Limpieza SUAVE (Solo quitamos espacios, NO minúsculas ni regex agresivo)
        # Los links NO se tocan con .lower() ni se les quitan signos
        df_final[col] = df_final[col].astype(str).str.strip()
        
        # Opcional: Si quedaron 'nan' como texto, los pasamos a 'unknown'
        df_final[col] = df_final[col].replace({'nan': 'unknown', 'None': 'unknown'})

print("✅ ¡Links restaurados! Verifica uno:")
print(df_final[cols_links[0]].iloc[0])
```

```python
import numpy as np

# --- 🚑 OPERACIÓN RESTAURACIÓN TOTAL (Desde el original) ---

# 1. Traemos la columna CRUDA desde el dataframe original (el sucio)
if 'Temperament' in df.columns:
    print("🔄 Recuperando datos originales...")
    df_final['Temperament'] = df['Temperament']

    # 2. Función ultra-sensible (No borra nada que parezca número)
    def limpieza_extrema(valor):
        texto = str(valor).lower().strip()
        if texto in ['nan', 'unknown', 'sin información', 'none', '']:
            return np.nan
        
        # Si tiene la barra (ej: 5/10), extrae el primer número
        if '/' in texto:
            parte = texto.split('/')[0].strip()
            try: return float(parte)
            except: return np.nan
            
        # Si es un número solo (ej: "8")
        try:
            # Quitamos cualquier letra que haya quedado por error
            solo_num = "".join(filter(str.isdigit, texto))
            return float(solo_num) if solo_num else np.nan
        except:
            return np.nan

    # 3. Aplicamos la limpieza
    df_final['Temperament'] = df_final['Temperament'].apply(limpieza_extrema)

    # 4. Calculamos el promedio REAL (solo de los que tienen datos)
    promedio_real = df_final['Temperament'].mean()
    
    if pd.isna(promedio_real):
        print("⚠️ Alerta: ¡Incluso en el original no hay números!")
        promedio_real = 5.0
    else:
        print(f"📊 Promedio real detectado: {round(promedio_real, 2)}")

    # 5. Rellenamos solo los huecos
    df_final['Temperament'] = df_final['Temperament'].fillna(round(promedio_real, 1))

print("\n✅ Resultado final (Primeros 10):")
print(df_final['Temperament'].head(10).tolist())
```
